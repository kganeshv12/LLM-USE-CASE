```python
from langchain.llms import LlamaCpp
from langchain.callbacks.manager import CallbackManager
from langchain.callbacks.streaming_stdout import StreamingStdOutCallbackHandler
from huggingface_hub import hf_hub_download
from langchain.chains.question_answering import load_qa_chain
```

```python

# Callbacks support token-wise streaming
callback_manager = CallbackManager([StreamingStdOutCallbackHandler()])
# Verbose is required to pass to the callback manager
```

```python

model_name_or_path = "TheBloke/Llama-2-13B-chat-GGUF"
model_basename = "llama-2-13b-chat.Q5_K_M.gguf" # the model is in bin format
```

```python

model_path = hf_hub_download(repo_id=model_name_or_path, filename=model_basename)
```

```python
n_gpu_layers = 40  # Change this value based on your model and your GPU VRAM pool.
n_batch = 256  # Should be between 1 and n_ctx, consider the amount of VRAM in your GPU.

# Loading model,
llm = LlamaCpp(
    model_path=model_path,
    max_tokens=256,
    n_gpu_layers=n_gpu_layers,
    n_batch=n_batch,
    callback_manager=callback_manager,
    n_ctx=1024,
    verbose=True,
)
```

```python
chain=load_qa_chain(llm, chain_type="stuff")
```

```python
query = "what are the various courses offered?"
docs=docsearch.similarity_search(query)
```

```python
chain.run(input_documents=docs, question=query)
```

 **Here's a detailed explanation of the code, line by line:**

**1. Importing Libraries:**

- **!pip install ...:** These lines install required libraries:
    - `langchain`: For building question-answering pipelines.
    - `pypdf`: For loading PDFs.
    - `unstructured`: For text processing.
    - `sentence_transformers`: For generating text embeddings.
    - `pinecone-client`: For interacting with Pinecone vectorstore.
    - `llama-cpp-python`: For using the Llama language model.
    - `huggingface_hub`: For accessing models from Hugging Face Hub.

**2. Loading PDF Content:**

- **loader = PyPDFLoader("/content/5th Sem RR Campus Course Information.pdf"):** Loads a PDF file using PyPDFLoader.
- **data = loader.load():** Extracts text data from the PDF.

**3. Splitting Text:**

- **text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=0):** Creates a text splitter that divides text into 500-character chunks without overlap.
- **docs = text_splitter.split_documents(data):** Splits the PDF text into chunks and stores them in the `docs` list.

**4. Setting Up Embeddings and Vectorstore:**

- **os.environ["HUGGINGFACEHUB_API_TOKEN"] = ...:** Sets a token for accessing Hugging Face Hub.
- **PINECONE_API_KEY = ..., PINECONE_API_ENV = ...:** Sets access keys for Pinecone vectorstore.
- **embeddings = HuggingFaceEmbeddings(model_name='sentence-transformers/all-MiniLM-L12-v2'):** Creates embeddings using a pre-trained sentence transformer model.
- **pinecone.init(api_key, environment):** Initializes Pinecone with provided credentials.
- **index_name = "mpdfchat":** Sets a name for the Pinecone index.
- **docsearch = Pinecone.from_texts([t.page_content for t in docs], embeddings, index_name=index_name):** Creates a Pinecone vectorstore, embeds the text chunks, and stores them in the index.

**5. Querying the Vectorstore:**

- **query = "what are the various courses offered?":** Sets a query to search for relevant text chunks.
- **docs = docsearch.similarity_search(query):** Retrieves text chunks most similar to the query from the vectorstore.

**6. Loading a Language Model (Llama):**

- **!CMAKE_ARGS="-DLLAMA_CUBLAS=on" ...:** Special installation instructions for Llama.
- **callback_manager = CallbackManager(...):** Sets up a callback manager for streaming model outputs.
- **model_path = hf_hub_download(repo_id, filename):** Downloads a Llama language model from Hugging Face Hub.
- **llm = LlamaCpp(model_path, ...):** Loads the Llama model with specified parameters.

**7. Creating a Question-Answering Chain:**

- **chain = load_qa_chain(llm, chain_type="stuff"):** Creates a question-answering chain using the loaded language model.

**8. Running Queries:**

- **query = ...:** Sets various query questions.
- **docs = docsearch.similarity_search(query):** Retrieves relevant text chunks from the vectorstore.
- **chain.run(input_documents=docs, question=query):** Runs the question-answering chain to generate answers using the retrieved text chunks and the query.

**9. Using a Different Language Model (Flan-T5):**

- **llm = HuggingFaceHub(repo_id="google/flan-t5-xxl", model_kwargs={"temperature":0.5, "max_length":512}):** Loads a Flan-T5 model from Hugging Face Hub.
- **chain = load_qa_chain(llm, chain_type="stuff"):** Creates a question-answering chain using the Flan-T5 model.
- **chain.run(input_documents=docs, question=query):** Runs the question-answering chain using the Flan-T5 model.
